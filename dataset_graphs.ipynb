```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("/WHO_Life_expectancy.csv")
df
```

```python
#Bar Chart
x = df['Region']
y = df["Infant_deaths"]
plt.bar(x,y)
plt.title("Region wise Infant_deaths")
plt.xlabel("Region")
plt.ylabel("Infant_deaths")
plt.show()
```

```python
#Line Chart

df_grouped = df.groupby('Year')['Life_expectancy'].mean()

plt.plot(df_grouped.index, df_grouped.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Average Life Expectancy')
plt.title('Average Life Expectancy Over Years')
plt.show()
```

```python
#Pie Chart
df_grouped = df.groupby('Region')['Alcohol_consumption'].mean()
# Remove missing values
df_grouped = df_grouped.dropna()
plt.pie(df_grouped.values, labels=df_grouped.index, autopct='%1.1f%%')
plt.title('Average Alcohol Consumption by Region')
plt.show()
```

```python
#stair chart
df_grouped = df.groupby('Year')['Adult_mortality'].mean().dropna()

plt.step(df_grouped.index, df_grouped.values, where='mid', marker='o')

plt.xlabel('Year')
plt.ylabel('Adult Mortality')
plt.title('Adult Mortality Over Years (Step Chart)')
plt.show()
```

```python
#Histogram
plt.hist(df['BMI'].dropna(), bins=8, edgecolor='black')

plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.title('BMI Distribution')

plt.show()
```

```python
#Missing Variables
df.isnull().sum()
```

```python
#Outliers
Q1 = df['Adult_mortality'].quantile(0.25)
Q3 = df['Adult_mortality'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Adult_mortality'] < lower_bound) | (df['Adult_mortality'] > upper_bound)]

print("Outliers:")
print(outliers)
```
